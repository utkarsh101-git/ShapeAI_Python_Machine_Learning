##Importing libraries

```python
import numpy as np
import pandas as pd
```

##Importing Dataframe

```python
from sklearn.datasets import load_boston
df = load_boston()
```

##Data Preprocessing

```python
dataset = pd.DataFrame(df.data, columns=df.feature_names)
```

```python
dataset["PRICE"] = df["target"]
print(dataset.head())
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
y = y.reshape(-1,1)
```

##Spliting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

##Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
X_test =  sc_X.transform(X_test)
y_test =  sc_y.transform(y_test)
```

##Training an Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
regressor_LR = LinearRegression()
regressor_LR.fit(X_train, y_train)
```

```python
y_pred_LR = regressor_LR.predict(X_test )
```

##Training Support Vector Regressor

```python
from sklearn.svm import SVR
regressor_SVR = SVR(kernel = 'rbf')
regressor_SVR.fit(X_train, y_train)
```

```python
y_pred_SVR = regressor_SVR.predict(X_test)
```

##Evaluating the results

```python
from sklearn.metrics import r2_score,mean_squared_error

print("Linear regression R-Squared at training: {}".format(r2_score(y_train, regressor_LR.predict(X_train)) ))
print("Linear regression R-Squared at testing: {}".format(r2_score(y_test, y_pred_LR) ))
print()
print("Support Vector regression R-Squared at training: {}".format(r2_score(y_train, regressor_SVR.predict(X_train)) ))
print("Support Vector regression R-Squared at testing: {}".format(r2_score(y_test, y_pred_SVR) ))

rmse1 = (np.sqrt(mean_squared_error(y_test, y_pred_LR)))
print()
print("The Linear Regression model performance for testing set")
print('RMSE is {}'.format(rmse1))

rmse2 = (np.sqrt(mean_squared_error(y_test, y_pred_SVR)))
print()
print("The Support Vector Regression model performance for testing set")
print('RMSE is {}'.format(rmse2))
```
